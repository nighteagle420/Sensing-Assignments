## Question 2(a)
We are importing pandas to read the csv file and plot the data using matplotlib. First we are plotting the sitting case.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to the CSV file
csv_file_path = 'sitting.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

plt.plot(df['time'], df['gFx'], label='G-Force X', color='blue')
plt.plot(df['time'], df['gFy'], label='G-Force Y', color='red')
plt.plot(df['time'], df['gFz'], label='G-Force Z', color='green')

# Adding labels and title
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('G-Force', fontsize=12)
plt.title('Time vs G-Force(Sitting)', fontsize=14)

# Adding a grid and a legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

Now we plot the standing case.

```python
csv_file_path = 'standing.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# New plot
plt.figure()
plt.plot(df['time'], df['gFx'], label='G-Force X', color='blue')
plt.plot(df['time'], df['gFy'], label='G-Force Y', color='red')
plt.plot(df['time'], df['gFz'], label='G-Force Z', color='green')

# Adding labels and title
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('G-Force', fontsize=12)
plt.title('Time vs G-Force(Standing)', fontsize=14)

# Adding a grid and a legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

Lastly, we are plotting the lying down case.

```python
csv_file_path = 'sleep.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

#New plot
plt.figure()
plt.plot(df['time'], df['gFx'], label='G-Force X', color='blue')
plt.plot(df['time'], df['gFy'], label='G-Force Y', color='red')
plt.plot(df['time'], df['gFz'], label='G-Force Z', color='green')

# Adding labels and title
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('G-Force', fontsize=12)
plt.title('Time vs G-Force(Lying down straight)', fontsize=14)

# Adding a grid and a legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
