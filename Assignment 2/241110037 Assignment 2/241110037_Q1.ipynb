## Question 1(b)
We are importing matplotlib and librosa for this code. My voice is saved in "audio.wav"

```python
import matplotlib.pyplot as plt
import librosa

def plot_specgram(audio_file):
    # Load the audio file
    y, sr = librosa.load(audio_file, sr=None)
    
    # Compute and plot the spectrogram
    plt.specgram(y, Fs=sr, cmap='inferno', NFFT=2048, noverlap=1024, scale='dB')
    
    # Add color bar
    plt.colorbar(label='Intensity (dB)')
    
    # Add title and labels
    plt.title('Spectrogram')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    
    # Show the plot
    plt.tight_layout()
    plt.show()

# Replace 'your_audio_file.wav' with the path to your audio file
plot_specgram('audio.wav')
```

# Observations
1) We see that frequency spans from 0 to 22000hz which is the human hearing range

2) The color scale on the right shows the intensity (or amplitude) of the frequencies in decibels (dB). Brighter yellow/orange areas indicate higher intensity, while the darker purple regions show lower intensity.

3) There is consistent frequency activity between 0 and 5,000 Hz, with periodic intensity peaks that are likely harmonics or repeated tones.

4) Above 10,000 Hz, the signal appears more fragmented, suggesting higher-frequency components but less intensity compared to the lower frequencies.

5) The bright vertical bands around 1 second, 2 seconds, and so on suggest periodic or pulsed components, indicating repetitive or rhythmic features in the signal.
